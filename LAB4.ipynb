```python
# Vigenere Cipher Implementation

# Function for encryption
def encrypt_vigenere(plaintext, key):
    ciphertext = ""

    # Convert plaintext and key to uppercase
    plaintext = plaintext.upper()
    key = key.upper()

    # Repeat key until it matches plaintext length
    key_repeated = ""
    for i in range(len(plaintext)):
        key_repeated += key[i % len(key)]

    # Encrypt each character
    for p, k in zip(plaintext, key_repeated):

        # Convert letters to numbers (A=0, B=1 ... Z=25)
        p_val = ord(p) - ord('A')
        k_val = ord(k) - ord('A')

        # Encryption formula:
        # C = (P + K) mod 26
        c_val = (p_val + k_val) % 26

        # Convert back to character
        ciphertext += chr(c_val + ord('A'))

    return ciphertext


# Function for decryption
def decrypt_vigenere(ciphertext, key):
    plaintext = ""

    ciphertext = ciphertext.upper()
    key = key.upper()

    # Repeat key
    key_repeated = ""
    for i in range(len(ciphertext)):
        key_repeated += key[i % len(key)]

    # Decrypt each character
    for c, k in zip(ciphertext, key_repeated):

        # Convert letters to numbers
        c_val = ord(c) - ord('A')
        k_val = ord(k) - ord('A')

        # Decryption formula:
        # P = (C - K + 26) mod 26
        p_val = (c_val - k_val + 26) % 26

        plaintext += chr(p_val + ord('A'))

    return plaintext


# Example from Lab
plaintext = "CYBERSECURITY"
key = "BEST"

ciphertext = encrypt_vigenere(plaintext, key)
print("Encrypted Text:", ciphertext)

decrypted = decrypt_vigenere(ciphertext, key)
print("Decrypted Text:", decrypted)
```

```python
# Playfair Cipher Implementation

# Create Playfair matrix
def generate_key_matrix(key):
    key = key.upper().replace('J', 'I')

    matrix = []
    used = set()

    # Add key characters
    for char in key:
        if char not in used and char.isalpha():
            matrix.append(char)
            used.add(char)

    # Add remaining alphabets
    for char in "ABCDEFGHIKLMNOPQRSTUVWXYZ":
        if char not in used:
            matrix.append(char)
            used.add(char)

    # Convert into 5x5 matrix
    return [matrix[i:i+5] for i in range(0, 25, 5)]


# Find position of a character in matrix
def find_position(matrix, char):
    for row in range(5):
        for col in range(5):
            if matrix[row][col] == char:
                return row, col


# Prepare plaintext
def prepare_text(text):
    text = text.upper().replace('J', 'I')

    prepared = ""
    i = 0

    while i < len(text):
        prepared += text[i]

        if i + 1 < len(text):
            if text[i] == text[i + 1]:
                prepared += 'X'
            else:
                prepared += text[i + 1]
                i += 1
        i += 1

    if len(prepared) % 2 != 0:
        prepared += 'X'

    return prepared


# Encryption
def encrypt_playfair(text, key):
    matrix = generate_key_matrix(key)
    text = prepare_text(text)

    cipher = ""

    for i in range(0, len(text), 2):
        a, b = text[i], text[i+1]

        r1, c1 = find_position(matrix, a)
        r2, c2 = find_position(matrix, b)

        # Same row
        if r1 == r2:
            cipher += matrix[r1][(c1 + 1) % 5]
            cipher += matrix[r2][(c2 + 1) % 5]

        # Same column
        elif c1 == c2:
            cipher += matrix[(r1 + 1) % 5][c1]
            cipher += matrix[(r2 + 1) % 5][c2]

        # Rectangle rule
        else:
            cipher += matrix[r1][c2]
            cipher += matrix[r2][c1]

    return cipher


# Decryption
def decrypt_playfair(cipher, key):
    matrix = generate_key_matrix(key)

    plaintext = ""

    for i in range(0, len(cipher), 2):
        a, b = cipher[i], cipher[i+1]

        r1, c1 = find_position(matrix, a)
        r2, c2 = find_position(matrix, b)

        # Same row
        if r1 == r2:
            plaintext += matrix[r1][(c1 - 1) % 5]
            plaintext += matrix[r2][(c2 - 1) % 5]

        # Same column
        elif c1 == c2:
            plaintext += matrix[(r1 - 1) % 5][c1]
            plaintext += matrix[(r2 - 1) % 5][c2]

        # Rectangle rule
        else:
            plaintext += matrix[r1][c2]
            plaintext += matrix[r2][c1]

    return plaintext


# Example from Lab
keyword = "MONARCHY"
plaintext = "INSTRUMENTS"

ciphertext = encrypt_playfair(plaintext, keyword)
print("Encrypted Text:", ciphertext)

decrypted = decrypt_playfair(ciphertext, keyword)
print("Decrypted Text:", decrypted)
```
