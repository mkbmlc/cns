```python
# RSA Encryption and Decryption Example

# Step 1: Choose two prime numbers
p = 61
q = 53

# Step 2: Calculate n
n = p * q

# Step 3: Calculate Euler's Totient Function
phi = (p - 1) * (q - 1)

# Step 4: Choose public exponent
e = 17

# Step 5: Find modular inverse of e modulo phi
# pow(e, -1, phi) gives the modular inverse
d = pow(e, -1, phi)

print("Public Key (e, n):", (e, n))
print("Private Key (d, n):", (d, n))

# Step 6: Plaintext message
message = 65

# Step 7: Encrypt
ciphertext = pow(message, e, n)

print("Encrypted Message:", ciphertext)

# Step 8: Decrypt
decrypted_message = pow(ciphertext, d, n)

print("Decrypted Message:", decrypted_message)
```

```python
# Diffie-Hellman Key Exchange

# Public parameters
q = 157      # Prime number
alpha = 5    # Primitive root

# Private keys chosen by users
XA = 15
XB = 27

# Step 1: Compute public values
YA = pow(alpha, XA, q)
YB = pow(alpha, XB, q)

print("YA =", YA)
print("YB =", YB)

# Step 2: Compute shared secret key

# Key computed by A
KA = pow(YB, XA, q)

# Key computed by B
KB = pow(YA, XB, q)

print("Key computed by A =", KA)
print("Key computed by B =", KB)

# Both keys should be equal
if KA == KB:
    print("Shared Secret Key =", KA)
else:
    print("Key Exchange Failed")
```
